```python
import pandas as pd

import datetime

import numpy as np

import matplotlib.pyplot as plt

from pandas.plotting import scatter_matrix

!pip install yfinance

import yfinance as yf

%matplotlib inline
```

# Tech

```python
#grabbing the data for 3 stocks over the course of 5 years
start = '2014-01-01'
end = '2019-1-01'
twt = yf.download('TWTR', start, end)
fb = yf.download('META', start, end)
goog = yf.download('GOOGL', start, end)
```

```python
#code will plot the opening prices of the 3 selected stocks
twt['Open'].plot(label = 'Twitter', figsize = (15,7))
fb['Open'].plot(label = "Facebook/Meta")
goog['Open'].plot(label = 'Google')
plt.title('Stock Prices of Twitter, Facebook, Google')
plt.legend()
```

```python
#this plots the volume of stock bought for the big 3
twt['Volume'].plot(label='Twitter',
                  figsize= (12,10))
fb['Volume'].plot(label='Facebook/Meta')
goog['Volume'].plot(label='Google')
plt.title('Volume of Stock Traded')
plt.legend()
```

```python
#Market Capitalisation
twt['MarktCap'] = twt['Open'] * twt['Volume']
fb['MarktCap'] = fb['Open'] * fb['Volume']
goog['MarktCap'] = goog['Open'] * goog['Volume']
twt['MarktCap'].plot(label = 'Twitter', figsize = (15,7))
fb['MarktCap'].plot(label = 'Facebook/Meta')
goog['MarktCap'].plot(label = 'Google')
plt.title('Market Cap')
plt.legend()
```

### Moving Averages

### Twitter 

```python
#Observing the volatility of the Twitter stock using the 50/200 day average
twt['MA50'] = twt['Open'].rolling(50).mean()
twt['MA200'] = twt['Open'].rolling(200).mean()
twt['Open'].plot(figsize = (15,7))
twt['MA50'].plot()
twt['MA200'].plot()
plt.legend()
```

### Facebook/Meta

```python
#Observing the volatility of the FB stock using the 50/200 day average
fb['MA50'] = fb['Open'].rolling(50).mean()
fb['MA200'] = fb['Open'].rolling(200).mean()
fb['Open'].plot(figsize = (15,7))
fb['MA50'].plot()
fb['MA200'].plot()
plt.legend()
```

### Google

```python
#Observing the volatility of the Google stock using the 50/200 day average
goog['MA50'] = goog['Open'].rolling(50).mean()
goog['MA200'] = goog['Open'].rolling(200).mean()
goog['Open'].plot(figsize = (15,7))
goog['MA50'].plot()
goog['MA200'].plot()
plt.legend()
```

### Cross Stock Opening Price Analysis

```python
data = pd.concat([twt['Open'],fb['Open'],goog['Open']],axis = 1)
data.columns = ['Twitter','FaceBook/Meta','Google']
scatter_matrix(data, figsize = (10,10), hist_kwds= {'bins':250})
```

### Volatiility Amongst the 3 Stocks

```python
#Volatility
twt['returns'] = (twt['Close']/twt['Close'].shift(1)) -1
fb['returns'] = (fb['Close']/fb['Close'].shift(1))-1
goog['returns'] = (goog['Close']/goog['Close'].shift(1)) - 1
twt['returns'].hist(bins = 100, label = 'Twitter', alpha = 0.5, figsize = (15,7))
fb['returns'].hist(bins = 100, label = 'Facebook/Meta', alpha = 0.5)
goog['returns'].hist(bins = 100, label = 'Google', alpha = 0.5)
plt.legend()
```
